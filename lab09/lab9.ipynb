```python
import sys
sys.path.append("stylegan2-ada-pytorch")  # Add the repo path

import torch
import numpy as np
import dnnlib  # Now it will work
import legacy  # Also included in the repo
import PIL.Image
import matplotlib.pyplot as plt


# model FFHQ of StyleGAN2-ADA
network_pkl = "https://nvlabs-fi-cdn.nvidia.com/stylegan2-ada-pytorch/pretrained/ffhq.pkl"

# load model
def load_model(network_pkl):
    print(f'Loading model from "{network_pkl}"...')
    with dnnlib.util.open_url(network_pkl) as f:
        G = legacy.load_network_pkl(f)['G_ema'].cuda()
    return G


def generate_latent_vector(G):
    z = torch.randn([1, G.z_dim]).cuda()
    return z

def manipulate_latent(G, z, direction, strength=1.5):
    modified_z = z + strength * direction
    return modified_z

# create the image from the latent vector
def generate_image(G, z):
    img = G(z, None)  
    img = (img + 1) * (255 / 2)  
    img = img.permute(0, 2, 3, 1).clamp(0, 255).to(torch.uint8)  
    img = img[0].cpu().numpy()
    return img

# Chạy chương trình chính
if __name__ == "__main__":
    G = load_model(network_pkl)

    # generate the original image
    z = generate_latent_vector(G)
    img_original = generate_image(G, z)

    # change the latent vector
    direction = torch.randn_like(z).cuda()  
    z_modified = manipulate_latent(G, z, direction, strength=2.0)
    img_modified = generate_image(G, z_modified)

    # display the images
    fig, ax = plt.subplots(1, 2, figsize=(10, 5))
    ax[0].imshow(img_original)
    ax[0].set_title("Original Face")
    ax[0].axis("off")

    ax[1].imshow(img_modified)
    ax[1].set_title("Modified Face")
    ax[1].axis("off")

    plt.show()
```

```python
import torch
import numpy as np
import matplotlib.pyplot as plt
from pytorch_pretrained_biggan import BigGAN, one_hot_from_names, truncated_noise_sample
import nltk

# Download WordNet data (if not already downloaded)
nltk.download('wordnet')

# Load the pre-trained BigGAN model
model = BigGAN.from_pretrained('biggan-deep-256').eval()

# Generate a random latent vector
z = torch.from_numpy(truncated_noise_sample(truncation=0.4, batch_size=1)).to(torch.float32)

# Create a class vector for "volcano"
class_vector = torch.from_numpy(one_hot_from_names(['volcano'], batch_size=1)).to(torch.float32)

# Generate image from BigGAN
with torch.no_grad():
    img_tensor = model(z, class_vector, 0.4)

# Normalize and convert to image
img_tensor = (img_tensor - img_tensor.min()) / (img_tensor.max() - img_tensor.min())  
img_array = (img_tensor.squeeze(0).cpu().numpy() * 255).astype(np.uint8)  

# Display image
plt.figure(figsize=(6, 6))
plt.imshow(img_array.transpose(1, 2, 0))
plt.axis("off")
plt.title("BigGAN Generated Artwork (Volcano)")
plt.show()
```
